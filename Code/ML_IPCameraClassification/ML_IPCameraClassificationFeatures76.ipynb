```python
#check for columns with NaN
inf_null_col_names = list()
for cni in col_names_onlyX:
    if df_cam[cni].isnull().values.any():
        print(cni,' :  isNull: ', df_cam[cni].isnull().values.any(), ' isNullCount: ',df_cam[cni].isnull().values.sum())
        if cni not in inf_null_col_names:
            inf_null_col_names.append(cni)
    c = np.isinf(df_cam[cni]).values.sum()
    if c > 0:
        print(cni, ' : infCount: ', c)
        if cni not in inf_null_col_names:
            inf_null_col_names.append(cni)
print('num cols with null or inf: ', len(inf_null_col_names))   
```

```python
#replace NaN with 0 for every column
df_cam.replace([np.inf, -np.inf], np.nan, inplace=True)

#### recheck for inf and nan
for cni in col_names_onlyX:
    if df_cam[cni].isnull().values.any():
        print(cni,' :  isNull: ', df_cam[cni].isnull().values.any(), ' isNullCount: ',df_cam[cni].isnull().values.sum())
        if cni not in inf_null_col_names:
            inf_null_col_names.append(cni)
    c = np.isinf(df_cam[cni]).values.sum()
    if c > 0:
        print(cni, ' : infCount: ', c)
        if cni not in inf_null_col_names:
            inf_null_col_names.append(cni)
print('num cols with null or inf: ', len(inf_null_col_names), ' col names: ', inf_null_col_names) 
print('df_labeled_ipcam_Netatmo.shape', df_labeled_ipcam_Netatmo.shape)
```
